```python
dna = load_dna_structure_files(input_path)
longest_strand, longest_strand_index = find_longest_strand(dna)

for iteration in range(num_iterations):
    mutants = []
    angles = []
    simulated_dna_list = [] 

    # Step 1: Find a valid point in the DNA structure
    point_pos = find_valid_point(dna, left_indices, right_indices, longest_strand)
    
    # Step 2: Remove three random staples within a sphere around the point
    new_dna_structures, removed_strands_info = remove_three_strands_in_sphere(dna, point_pos, sphere_radius)
    
    # Determine the number of mutants based on the number of new structures created
    num_mutants = len(new_dna_structures)
    print(f"Iteration {iteration}: Generated {num_mutants} new structures.")
    
    export_paths = export_dna_structures(new_dna_structures, base_path)

    # Step 3: Simulate each modified structure
    for export_path in export_paths:
        structure_id = export_path['structure_id']
        print(f"Starting simulations for structure {structure_id}...")
        run_simulations_for_structure(structure_id, base_path, sim_base_path, rel_parameters, eq_parameters, prod_parameters)
        print(f"Simulations for structure {structure_id} completed.")

    # Step 4: Measure the angle at the joint for each mutant after simulation
    for export_path in export_paths:
        structure_id = export_path['structure_id']
        simulated_dna = load_simulated_structure(structure_id, sim_base_path)
        simulated_dna_list.append(simulated_dna)  
        point_pos, bend_angle = find_bend_angle(simulated_dna, left_indices, right_indices, longest_strand)
        mutants.append(simulated_dna)
        angles.append(bend_angle)

    # Step 7: Evaluate fitness
    fitness_scores = evaluate_fitness(angles, desired_angle, tolerance)
    
    # Select the best mutant based on fitness scores
    best_mutant_index = np.argmin(fitness_scores)
    best_mutant_structure = mutants[best_mutant_index]
    best_angle = angles[best_mutant_index]

    print(f"Iteration {iteration + 1}: Best angle = {best_angle} degrees (Structure {best_mutant_index}).")

    # Update the structure with the best mutant
    dna = best_mutant_structure

    # Output progress
    print(f"Iteration {iteration + 1}/{num_iterations}: Best angle = {best_angle} degrees")

    # Check if the best angle is within the desired tolerance
    if abs(best_angle - desired_angle) <= tolerance:
        print("Desired angle achieved within tolerance.")
        break

# Save the final optimized structure
export_dna_structures([best_mutant_structure], save_path)
print(f"Final optimized structure saved to {save_path}.")
```
