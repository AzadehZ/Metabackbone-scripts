```python
# Function to create new left and right indices after removing staples
def update_left_right_indices(mutants, original_structure, left_indices, right_indices, removed_staples_info):
    all_new_left_indices = []
    all_new_right_indices = []

    for structure, removed_strands in zip(mutants, removed_staples_info):
        new_left_indices = []
        new_right_indices = []

        removed_bases_indices = []
        for strand in removed_strands:
            strand = int(strand)  # Ensure strand is of correct type
            for base in original_structure.strands[strand]:
                removed_bases_indices.append(base.uid)

        max_removed_index = max(removed_bases_indices) if removed_bases_indices else -1

        for index in left_indices:
            if index < max_removed_index:
                new_left_indices.append(index)
            else:
                new_left_indices.append(index - len(removed_bases_indices))

        for index in right_indices:
            if index < max_removed_index:
                new_right_indices.append(index)
            else:
                new_right_indices.append(index - len(removed_bases_indices))

        all_new_left_indices.append(new_left_indices)
        all_new_right_indices.append(new_right_indices)

    return all_new_left_indices, all_new_right_indices
```
